```
import cv2
import matplotlib.pyplot as plt
import numpy as np
from keras.datasets import cifar10
(cx_train, cy_ftrain), (cx_test, cy_test) = cifar10.load_data()
```

## cx_train[40] 영상을 img 변수로 저장하고 plt.imshow 함수를 이용하여 가시화기 위한 코드를 작성하시오

```
img = cx_train[40]
height, width, channel = img.shape
plt.imshow(img)
print(height, width, channel)
```

##칼라 영상으로부터 r channel, g channel, b channel 별로 구분한 후 각 채널별 256 단계로 히스트로그램 및 normalized histogram을 구하기 위한 함수를 histogram_256, normal_histogram_256의 이름을 가지도록 코드를 작성한 후 각 채널별 히스토그램을 가시화 하시오.

```
R_channel = img[:, :, 0]
G_channel = img[:, :, 1]
B_channel = img[:, :, 2]

def histogram_256(img):
  bins = np.zeros(256, np.int32)
  for i in range(0, height):
    for j in range(0, width):
      bins[img[i][j]] += 1

  return bins

R_bins = histogram_256(R_channel)
G_bins = histogram_256(G_channel)
B_bins = histogram_256(B_channel)

plt.subplot(3,1,1)
plt.plot(R_bins, 'red')
plt.title('R_channel_histogram')
plt.show()

plt.subplot(3,1,2)
plt.plot(G_bins, 'green')
plt.title('G_channel_histogram')
plt.show()

plt.subplot(3,1,3)
plt.plot(B_bins, 'blue')
plt.title('B_channel_histogram')
plt.show()

def normal_histogram_256(img):
  bins = np.zeros(256, np.int32)
  for i in range(0, height):
    for j in range(0, width):
      bins[img[i][j]] += 1

  bins = bins/(width*height)  #normalized histogram
  return bins

N_R_bins = normal_histogram_256(R_channel)
N_G_bins = normal_histogram_256(G_channel)
N_B_bins = normal_histogram_256(B_channel)

plt.subplot(3,1,1)
plt.plot(N_R_bins, 'red')
plt.title('R_channel_normalized_histogram')
plt.show()

plt.subplot(3,1,2)
plt.plot(N_G_bins, 'green')
plt.title('G_channel_normalized_histogram')
plt.show()

plt.subplot(3,1,3)
plt.plot(N_B_bins, 'blue')
plt.title('B_channel_normalized_histogram')
plt.show()
```

##각 r channel, g channel, b channel별 16 단계로 단계로 히스트로그램 및 normalized histogram을 구하기 위한 함수를 histogram_256, normal_histogram_256의 이름을 가지도록 코드를 작성한 후 각 채널별 히스토그램을 가시화 하시오

```

def histogram_16(img):
  bins = np.zeros(16, np.int32)
  for i in range(0, height):
    for j in range(0, width):
      bins[int(img[i][j]/16)] += 1

  return bins

R_bins = histogram_16(R_channel)
G_bins = histogram_16(G_channel)
B_bins = histogram_16(B_channel)

plt.subplot(3,1,1)
plt.plot(R_bins, 'red')
plt.title('R_channel_histogram')
plt.show()

plt.subplot(3,1,2)
plt.plot(G_bins, 'green')
plt.title('G_channel_histogram')
plt.show()

plt.subplot(3,1,3)
plt.plot(B_bins, 'blue')
plt.title('B_channel_histogram')
plt.show()

def normal_histogram_16(img):
  bins = np.zeros(16, np.int32)
  for i in range(0, height):
    for j in range(0, width):
      bins[int(img[i][j]/16)] += 1

  bins = bins/(width*height)  #normalized histogram
  return bins

N_R_bins = normal_histogram_16(R_channel)
N_G_bins = normal_histogram_16(G_channel)
N_B_bins = normal_histogram_16(B_channel)

plt.subplot(3,1,1)
plt.plot(N_R_bins, 'red')
plt.title('R_channel_normalized_histogram')
plt.show()

plt.subplot(3,1,2)
plt.plot(N_G_bins, 'green')
plt.title('G_channel_normalized_histogram')
plt.show()

plt.subplot(3,1,3)
plt.plot(N_B_bins, 'blue')
plt.title('B_channel_normalized_histogram')
plt.show()
```
